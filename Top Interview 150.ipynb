### Merge_Sorted_Array

```python
def merge(nums1, m: int, nums2, n: int) -> None:
    idx1 = 0
    idx2 = 0
    sorted_arr = []
    while idx1 < m and idx2 < n:
        if nums1[idx1] < nums2[idx2]:
            sorted_arr.append(nums1[idx1])
            idx1 += 1
        else:
            sorted_arr.append(nums2[idx2])
            idx2 += 1

    
    while idx2 < n:
        sorted_arr.append(nums2[idx2])
        idx2 += 1
    
    while idx1 < m:
        sorted_arr.append(nums1[idx1])
        idx1 += 1

    for idx in range(len(nums1)):
        nums1[idx] = sorted_arr[idx]
```

```python
nums1 = [1,2,3,0,0,0]
m = 3
nums2 = [2,5,6]
n = 3

merge(nums1, m, nums2, n)
nums1
```

#### Another Solution

```python
def merge(nums1, m: int, nums2, n: int) -> None:
    idx1 = m - 1
    idx2 = n - 1
    next_biggest = m + n - 1
    while idx2 >= 0:
        if idx1 >= 0 and nums1[idx1] > nums2[idx2]:
            nums1[next_biggest] = nums1[idx1]
            idx1 -= 1
        else:
            nums1[next_biggest] = nums2[idx2]
            idx2 -= 1

        next_biggest -= 1
```

```python
nums1 = [1,2,3,0,0,0]
m = 3
nums2 = [2,5,6]
n = 3

merge(nums1, m, nums2, n)
nums1
```

### Remove_Element

```python
def removeElement(nums, val: int) -> int:
    index = 0
    for i in range(len(nums)):
        if nums[i] != val:
            nums[index] = nums[i]
            index += 1

    return index
```

```python
nums = [4, 5]

#nums = [2]
removeElement(nums, 4)
```

### Remove_Duplicates_from_Sorted_Array_Easy

```python
def removeDuplicates(nums) -> int:
    unique_arr = []
    k = 0
    for num in nums:
        if num not in unique_arr:
            unique_arr.append(num)
            k += 1
    
    for i in range(k):
        nums[i] = unique_arr[i]
        
    return k
```

```python
arr = [1, 1, 2]
removeDuplicates(arr)
```

```python
def removeDuplicates(nums) -> int:
    j = 1
    for i in range(1, len(nums)):
        if nums[i] != nums[j-1]:
            nums[j] = nums[i]
            j += 1

    return j

```

```python
arr = [1, 1, 2]
removeDuplicates(arr)
```

### Remove_Duplicates_from_Sorted_Array_II

```python
def removeDuplicates(nums) -> int:
    arr_size = len(nums)
    if arr_size > 2:
        j = 2
        for i in range(2, arr_size):
            if nums[j-2] != nums[i]:
                nums[j] = nums[i]
                j += 1

        arr_size = j
    return arr_size
```

```python
nums = [0, 0, 1, 1, 1, 1, 2, 3, 3]
removeDuplicates(nums)
```

### Majority_Element

```python
def majorityElement(nums) -> int:
    nums_count = {}
    for num in nums:
        nums_count[num] = nums_count.get(num, 0) + 1

    
    return list(filter(lambda key:nums_count[key] > (len(nums) / 2), nums_count))[0]
```

```python
# better solution
def majorityElement(nums) -> int:
    nums_count = {}
    for num in nums:
        nums_count[num] = nums_count.get(num, 0) + 1
    return sorted(nums_count, key=lambda key:nums_count[key])[-1]
```

```python
# Moore Voting Algorithm
def majorityElement(nums) -> int:
    num_vote = 0
    candidate = 0
    for num in nums:
        if num_vote == 0:
            candidate = num
            
        if candidate == num:
            num_vote += 1
        else:
            num_vote -= 1
            
    return candidate
```

```python
nums = [2,2,1,1,1,2,2]
#nums = [3, 2, 3]
majorityElement(nums)
```

### Rotate_Array

```python
def rotate(nums, k: int) -> None:
    """
    Do not return anything, modify nums in-place instead.
    """
    arr_size = len(nums)
    # If k = len(nums), we get the same array, so we use k % len(nums)
    # This ensures that if k is greater than the size of the array, we only rotate by the remainder.
    k = k % arr_size
    ans = nums[-k:] + nums[:-k]
    for i in range(len(nums)):
        nums[i] = ans[i]
```

```python
nums = [1,2,3,4,5,6,7]
#nums = [-1,-100,3,99]
#nums = [1, 2]
k=3
rotate(nums, k)
nums
```

### Best_Time_to_Buy_and_Sell_Stock

```python
def maxProfit(prices) -> int:
    max_profit = 0
    buy_price = prices[0]
    for sell_price in prices[1:]:
        if sell_price < buy_price:
            buy_price = sell_price

        max_profit = max(max_profit, sell_price - buy_price)
    return max_profit
```

```python
prices = [7,1,5,3,6,4]
#prices = [7,6,4,3,1]
#prices = [7, 3, 10, 1, 6, 4]
maxProfit(prices)
```

### Best_Time_to_Buy_and_Sell_Stock_II

```python
def maxProfit(prices) -> int:
    max_profit = 0
    buy_price = prices[0]
    for price in prices[1:]:
        if price > buy_price:
            max_profit += (price - buy_price)

        buy_price = price
    return max_profit
```

```python
def maxProfit(prices) -> int:
    max_profit = 0
    for i in range(1, len(prices)):
        max_profit += max(0, prices[i] - prices[i-1])
    return max_profit
```

```python
prices = [7,1,5,3,6,4]
#prices = [1,2,3,4,5]
#prices = [7,6,4,3,1]
maxProfit(prices)
```
